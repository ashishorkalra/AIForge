```python
text = "Sure! I can help with that. when are you available for the visit?"
output_file = "output.mp3"
```

```python
import edge_tts
import asyncio

async def generate_speech(
    text: str,
    output_file: str = "output.mp3",
    rate: str = "+0%",
    volume: str = "+0%",
    pitch: str = "+0Hz",
    voice: str = "en-US-MichelleNeural"
):
    """Synthesize speech using Microsoft Edge TTS with adjustable parameters.
    
    Args:
        text: Input text to synthesize
        output_file: Output filename (default: output.mp3)
        rate: Speaking rate adjustment (-100% to +100%)
        volume: Volume adjustment (-100% to +100%)
        pitch: Pitch adjustment (-100% to +100%)
        voice: Voice model (default: en-US-JennyNeural)
    """
    tts = edge_tts.Communicate(
        text=text,
        voice=voice,
        rate=rate,
        volume=volume,
        pitch=pitch
    )
    await tts.save(output_file)
```

```python
await generate_speech(text=text)
```
